```python
import numpy as np
import pandas as pd
parent_codes = pd.read_excel('2017_to_2012_NAICS.xlsx')
parent_codes = parent_codes.drop(parent_codes.index[0])
parent_codes
```

```python
# creating node properties table for Neo4j insertion
node_properties= pd.DataFrame(columns=['Node_Primary_Key','Node_Type_Key','Property_Name','Property'])
node_properties
```

```python
# creating edge properties table for Neo4j insertion
edge_properties = pd.DataFrame(columns = ['Edge_Primary_Key','Edge_Type_Key','Property_Name','Property'])
edge_properties
```

```python
for i in range(len(parent_codes)):
    t = {'Node_Primary_Key': [i, i, i], 'Node_Type_Key': [6, 6, 6], 
         'Property_Name' : ['Code', 'Code_Description','Quantity_NAICS_Industries'], 
         'Property' : [parent_codes['Sector'][i+1], parent_codes['Name'][i+1], parent_codes['NAICS Industries (5-digit)'][i+1]]}
    temp_node = pd.DataFrame(data = t)
    node_properties = node_properties.append(temp_node, ignore_index = True)
    
count = i
node_properties
```

```python
codes = pd.read_pickle("company_codes.pkl")
codes["Description"] = np.nan
codes
```

```python
codes['Code'][13] = 334111
```

```python
kid_codes = pd.read_excel('2017_to_2012_NAICS.xlsx', sheet_name = 1, header = 2)
for i in range(len(codes)):
    for j in range(len(kid_codes)):
        if kid_codes['2017 NAICS Code'][j] == codes['Code'][i]:
            codes['Description'][i] = kid_codes['2017 NAICS Title'][j]
```

```python
codes
```

```python
#Creating Nodes
edge_count = 0
for i in range(len(codes)):
    if codes['Description'][i] != np.nan:
        count += 1
        t = {'Node_Primary_Key': [count, count, count], 'Node_Type_Key': [6, 6, 6], 
             'Property_Name' : ['Name', 'Code', 'Code_Description'], 
             'Property' : [codes['Company'][i] + ' NAICS', codes['Code'][i], codes['Description'][i]]}
        temp_node = pd.DataFrame(data = t)
        node_properties = node_properties.append(temp_node, ignore_index = True)
        
        j = 0
        while j <= 1:
            if edge_count < 17:
                tt = 5
            else:
                tt = 6
            e = {'Edge_Primary_Key': [edge_count], 'Edge_Type_Key': [tt], 
                 'Property_Name' : ['Name'], 
                 'Property' : ['Code_Relation']}
            temp_edge = pd.DataFrame(data = e)
            edge_properties = edge_properties.append(temp_edge, ignore_index = True)
            edge_count += 1
            j += 1
```

```python
# creating connections table for Neo4j insertion
connections= pd.DataFrame(columns=['Node_Property_Key','Edge_Property_Key','Node_Property_Key_2'])
```

```python
for i in range(len(codes)):
    for j in range(len(node_properties)):
        if len(str(node_properties['Property'][j])) > 6:
            if node_properties['Property'][j][:-6] == codes['Company'][i]:
                node_a = node_properties['Node_Primary_Key'][j]
    try:         
        node_b = node_properties.loc[node_properties['Property'] == int(str(codes['Code'][i])[:2])]['Node_Primary_Key'].reset_index()['Node_Primary_Key'][0]
    except:
        for j in range(len(node_properties)):
            if isinstance(node_properties['Property'][i], str):
                if node_properties['Property'][i][:2] == str(codes['Code'][i])[:2]:
                    node_b = node_properties['Node_Primary_Key'][j]
                elif node_properties['Property'][i][-2:] == str(codes['Code'][i])[:2]:
                    node_b = node_properties['Node_Primary_Key'][j]
    
    c = {'Node_Property_Key':[node_a], 'Edge_Property_Key': [i], 'Node_Property_Key_2': [node_b]}
    temp_conn = pd.DataFrame(data = c)
    connections = connections.append(temp_conn, ignore_index = True)
```

```python
edge_properties.to_pickle("NAICS_edge_properties.pkl")
node_properties.to_pickle("NAICS_node_properties.pkl")
connections.to_pickle("NAICS_conn_properties.pkl")
```
