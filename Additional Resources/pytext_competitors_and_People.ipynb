```python
# Loading in json
import pandas as pd
import json
from pytextrank import top_keywords_sentences
from gensim.parsing.preprocessing import remove_stopwords
from gensim.models.phrases import Phrases, Phraser
from nltk.tokenize.treebank import TreebankWordDetokenizer
from gensim.utils import simple_preprocess
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
with open('new_competitors.json') as json_file:
    data = json.load(json_file)
```

```python
import numpy as np
```

```python
company_list = list()
for i in range(len(data)):
    company_list.append(data[i]['Title'])
```

```python
company_list
```

```python
competitor_people = pd.read_csv('competitor people.csv')
competitor_people
```

```python
competitor_table = pd.read_csv('competitors with titles.csv')
competitor_table['Company'][4] = 'VMware'
competitor_table['Company'][12] = 'Netgear'
competitor_table['Number_of_Founders'] = competitor_people['Number of founders']
competitor_table
```

```python
competitor_ed = pd.read_csv('competitors edges.csv')
competitor_ed['Company'][4] = 'VMware'
competitor_ed['Company'][12] = 'Netgear'
competitor_ed
```

```python
# creating node properties table for Neo4j insertion
node_properties= pd.DataFrame(columns=['Node_Primary_Key','Node_Type_Key','Property_Name','Property'])
node_properties
```

```python
# creating edge properties table for Neo4j insertion
edge_properties = pd.DataFrame(columns = ['Edge_Primary_Key','Edge_Type_Key','Property_Name','Property'])
edge_properties
```

```python
for i in range(len(company_list)):
    t = {'Node_Primary_Key' : [i], 'Node_Type_Key' : [2], 'Property_Name' : ['Name'], 'Property' : [company_list[i]]}
    temp_node = pd.DataFrame(data = t)
    node_properties = node_properties.append(temp_node, ignore_index = True)
```

```python
node_properties
```

```python
for i in range(1,len(company_list)):
    pkey = node_properties.loc[node_properties['Property'] == company_list[i]]['Node_Primary_Key'][i]
    value = competitor_table.loc[competitor_table['Company'] == company_list[i]].reset_index()
    ed = competitor_ed.loc[competitor_ed['Company'] == company_list[i]].reset_index()
    if value[competitor_table.columns[7]][0] == 'Unknown':
        value[competitor_table.columns[7]][0] = np.nan
    else:
        value[competitor_table.columns[7]][0] = float(value[competitor_table.columns[7]][0]) * 1e9
    
    t = {'Node_Primary_Key' : [pkey, pkey, pkey, pkey, pkey, pkey, pkey], 'Node_Type_Key' : [2, 2, 2, 2, 2, 2, 2], 
         'Property_Name' : [competitor_table.columns[2], competitor_table.columns[3], competitor_table.columns[4], 
                            competitor_table.columns[5], competitor_table.columns[6], competitor_table.columns[7], 
                            competitor_table.columns[8]], 
         'Property' : [value[competitor_table.columns[2]][0], value[competitor_table.columns[3]][0], int(value[competitor_table.columns[4]][0]), 
                       value[competitor_table.columns[5]][0], int(value[competitor_table.columns[6]][0]), value[competitor_table.columns[7]][0],
                       value[competitor_table.columns[8]][0]]}
    temp = pd.DataFrame(data = t)
    node_properties = node_properties.append(temp, ignore_index = True)
    
    e = {'Edge_Primary_Key': [i-1, i-1], 'Edge_Type_Key': [2, 2], 'Property_Name': [competitor_table.columns[1], 'Reason'], 
         'Property': [float(value[competitor_table.columns[1]][0]), ed['Unnamed: 2'][0]]}
    temp_edge = pd.DataFrame(data = e)
    edge_properties = edge_properties.append(temp_edge, ignore_index = True)
```

```python
t = {'Node_Primary_Key' : [0, 0, 0, 0, 0, 0, 0], 'Node_Type_Key' : [2, 2, 2, 2, 2, 2, 2], 
     'Property_Name' : [competitor_table.columns[2], competitor_table.columns[3], 
                        competitor_table.columns[4], competitor_table.columns[5], 
                        competitor_table.columns[6], competitor_table.columns[7],
                        competitor_table.columns[8]], 
     'Property' : ['Public', 'NASDAQ', 35, 'California', 74200, 49.33 * 1e9, 2]}

temp = pd.DataFrame(data = t)
node_properties = node_properties.append(temp, ignore_index = True)
```

```python
competitor_people = pd.read_csv('competitor people.csv')
competitor_people
```

```python
t = {'Company' : ['Cisco Systems'], 'CEO' : ['Chuck Robbins'], 'Highest Degree' : ['B.S.'], 'CEO gender': ['Male'] ,
     'Number of founders': [2]}
temp = pd.DataFrame(data = t)
competitor_people = competitor_people.append(temp, ignore_index = True)
competitor_people['Company'][4] = 'VMware'
competitor_people['Company'][12] = 'Netgear'
competitor_people
```

```python
# creating node properties table for Neo4j insertion
node_properties_p = pd.DataFrame(columns=['Node_Primary_Key','Node_Type_Key','Property_Name','Property'])
node_properties_p
```

```python
# creating edge properties table for Neo4j insertion
edge_properties_p = pd.DataFrame(columns = ['Edge_Primary_Key','Edge_Type_Key','Property_Name','Property'])
edge_properties_p
```

```python
for i in range(len(company_list)):
    ind = competitor_people.index[competitor_people['Company'] == company_list[i]][0]
    t = {'Node_Primary_Key' : [i, i, i], 'Node_Type_Key' : [1, 1, 1], 
         'Property_Name' : ['Name', 'Highest Degree', 'Gender'], 
         'Property' : [competitor_people['CEO'][ind], competitor_people['Highest Degree'][ind],
                       competitor_people['CEO gender'][ind]]}
    temp_node = pd.DataFrame(data = t)
    node_properties_p = node_properties_p.append(temp_node, ignore_index = True)
    
    e = {'Edge_Primary_Key': [i], 'Edge_Type_Key': [1], 'Property_Name': ['Position'], 
         'Property': ['CEO']}
    temp_edge = pd.DataFrame(data = e)
    edge_properties_p = edge_properties_p.append(temp_edge, ignore_index = True)
```
